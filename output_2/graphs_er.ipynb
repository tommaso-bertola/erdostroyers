```r
df_sur_conn<-read.csv('er/er_1000n_11lognnp_40means_zoom.csv')
df_er_1_n <- read.csv("er/er_1000n_1np_40means_zoom.csv")
df_er_11_n<-read.csv('er/er_1000n_11np_40means_zoom.csv')
df_er_13_n<-read.csv('er/er_1000n_13lognnp_40means_zoom.csv')
```

```r
df_er_13_n['p']<-'1.3 log(n)/n'
df_sur_conn['p']<-'1.1 log(n)/n'
df_er_1_n['p']<-'1/n'
df_er_11_n['p']<-'1.1/n'

df<-rbind(df_er_13_n, df_sur_conn, df_er_1_n, df_er_11_n)
```

```r
options(repr.plot.width = 10, repr.plot.height = 7)
plot_er_z <- ggplot(df, aes(x = alpha,
                            y = g_ratio)) +
geom_point(aes(col = type,
               shape=p),
           size = 4) +
geom_line(aes(col = type,
              shape=p),
          linewidth = 0.7,
          linetype=5) +
geom_linerange(aes(x = alpha,
                   ymin = g_ratio - g_ratio_sd,
                   ymax = g_ratio + g_ratio_sd,
                   col = type),
               position = position_dodge(0.005)) +
ylim(0, 1) +
xlim(0, 0.2) +
labs(x = expression(paste("Parameter ", alpha)),
     y = "G ratio",
     col = "Attack protocol",
     shape='Wiring probablity',
     title = expression(paste("Erdos-Renyi model"))) +
theme_bw() +
theme(text = element_text(size = 20))

plot_er_z

# ggsave(plot = plot_er_z, filename = 'er_zoom_1000n_11lognnp_40means.pdf',
# width = 10, height = 7, scale = 0.75)

ggsave(plot_er_z, filename = "er_zoom_1000n_40means.pdf", width = 10, height = 7,
    scale = 0.75)
```

```r
df_sbm <- read.csv("sbm//sbm_1000n_40means_new.csv")

plot_sbm <- ggplot(df_sbm, aes(x = alpha,
                            y = g_ratio)) +
geom_point(aes(col = type,
               ),
           size = 4) +
geom_line(aes(col = type,
              ),
          linewidth = 0.7,
          linetype=5) +
geom_linerange(aes(x = alpha,
                   ymin = g_ratio - g_ratio_sd,
                   ymax = g_ratio + g_ratio_sd,
                   col = type),
               position = position_dodge(0.005)) +
ylim(0, 1) +
xlim(0, 0.5) +
labs(x = expression(paste("Parameter ", alpha)),
     y = "G ratio",
     col = "Attack protocol",
     shape='Wiring probablity',
     title = expression(paste("Stochastic block model"))) +
theme_bw() +
theme(text = element_text(size = 20))

plot_sbm

ggsave(plot = plot_sbm, filename = "sbm_1000n_40means.pdf", width = 10, height = 7,
    scale = 0.75)
```

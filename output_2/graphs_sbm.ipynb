```r
df_sbm <- read.csv("sbm//sbm_1000n_40means_new.csv")

plot_sbm <- ggplot(df_sbm, aes(x = alpha, y = g_ratio, col = type)) + geom_point() +
    geom_line(alpha = 0.5, linewidth = 1.2) + geom_linerange(aes(x = alpha, ymin = g_ratio -
    g_ratio_sd, ymax = g_ratio + g_ratio_sd, col = type), position = position_dodge(0.03)) +
    ylim(0, 1) + labs(x = expression(paste("Parameter ", alpha)), y = "G ratio",
    col = "Attack protocol", title = expression(paste("Stochastic block model, 8 patches"))) +
    theme_bw() + xlim(0, 0.8)

plot_sbm

ggsave(plot = plot_sbm, filename = "sbm_1000n_40means.pdf", width = 10, height = 7,
    scale = 0.75)
```
